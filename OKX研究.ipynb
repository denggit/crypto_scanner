```python
import os
import sys
from dotenv import load_dotenv
from okx_api.client import OKXClient
from okx_api.market_data import MarketDataRetriever

load_dotenv()
```

```python
client = OKXClient(
    api_key=os.getenv("OK-ACCESS-KEY"),
    api_secret=os.getenv("OK-ACCESS-SECRET"),
    passphrase=os.getenv("OK-ACCESS-PASSPHRASE")
)
```

```python
# Initialize market data retriever
market_data = MarketDataRetriever(client)

# Get all spot tickers
tickers = market_data.get_all_tickers('SWAP')

# Get specific ticker
btc_ticker = market_data.get_ticker_by_symbol('BNB-USDT-SWAP')

# Get order book
order_book = market_data.get_order_book('BNB-USDT-SWAP', sz=10)

# Get kline data
klines = market_data.get_kline('BNB-USDT-SWAP', bar='1m', limit=100)

# Get kline data with MA
klines_ma = market_data.get_kline_with_ma('BNB-USDT-SWAP', bar='1m', limit=100, ma_periods=[5, 20, 60, 200])
```

```python
for ticker in tickers:
    if "BNB" in ticker.instId:
        print(ticker.instId)
```
