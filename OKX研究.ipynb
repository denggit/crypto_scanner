```python
import os
import sys
from dotenv import load_dotenv
from okx_api.client import OKXClient
from okx_api.market_data import MarketDataRetriever
from tools.market_scanner import CryptoScanner

load_dotenv()
```

```python
client = OKXClient(
    api_key=os.getenv("OK-ACCESS-KEY"),
    api_secret=os.getenv("OK-ACCESS-SECRET"),
    passphrase=os.getenv("OK-ACCESS-PASSPHRASE")
)

# Initialize market data retriever
market_data = MarketDataRetriever(client)

# Initialize crypto scanner
crypto_scanner = CryptoScanner(client)
```

```python
bullish_coins_1 = crypto_scanner.scan_ma_alignment(bar='5m', ma_periods=[5, 20, 60, 120])
bullish_coins_2 = crypto_scanner.scan_ma_convergence_breakout(bar='5m')
```

```python
print(f"total: {len(bullish_coins_1)}")
for coin in bullish_coins_1:
    print(coin.get("symbol"))
```
